```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data=pd.read_csv('C:\\Users\\Lenovo\\Documents\\data science\\movie_dataset.csv')
data.head(2)
```

```python
data.shape
```

```python
data.columns
```

```python
features=['keywords','cast','director','genres']
```

```python
for feature in features:
    data[feature]=data[feature].fillna('')
```

```python
def combine_features(row):
    try:
        return row['keywords']+''+row['cast']+''+row['director']+''+row['genres']
    except:
        print('error',row)
```

```python
data['combinedfeatures']=data.apply(combine_features,axis=1)
```

```python
data.head(1)
```

```python
count_matrix=cv.fit_transform(data['combinedfeatures'])
print(count_matrix.toarray())
```

```python
cosine_sim=cosine_similarity(count_matrix)
print(cosine_sim)
```

```python
movie_user_likes='Avatar'
```

```python
def get_index_from_title(title):
    return data[data.title ==title]['index'].values[0]
```

```python
def get_title_from_index(index):
    return data[data.index==index]['title'].values[0]
```

```python
movie_index=get_index_from_title(movie_user_likes)
```

```python
similar_movies=list(enumerate(cosine_sim[movie_index]))
```

```python
sorted_similar_movies=sorted(similar_movies,key=lambda x:x[1],reverse=True)
```

```python
print(sorted_similar_movies)
```
